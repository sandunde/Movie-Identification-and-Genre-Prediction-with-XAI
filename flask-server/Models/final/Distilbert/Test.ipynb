```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from transformers import pipeline

data = pd.read_csv('/Users/sandundesilva/Documents/4th year/Research Project/UI/findMyFilm/flask-server/Models/final/GenreDataset/Final_movie_train.csv')

X = data['subtitles']
y_true = data['genre']

X_train, X_test, y_train, y_test = train_test_split(X, y_true, test_size=0.2, random_state=42)

# Initialize the text classification pipeline with maximum sequence length
classifier = pipeline('text-classification', model='nickmuchi/distilbert-base-movie-genre-prediction', tokenizer='nickmuchi/distilbert-base-movie-genre-prediction', framework='pt', max_length=512)

# Predict labels for the testing set
y_pred = classifier(X_test.tolist())

# Extract predicted labels from the pipeline output
y_pred_labels = [prediction['label'] for prediction in y_pred]

# Calculate evaluation metrics
conf_matrix = confusion_matrix(y_test, y_pred_labels)
report = classification_report(y_test, y_pred_labels)
accuracy = accuracy_score(y_test, y_pred_labels)

# Print results
print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(report)
print("\nAccuracy:", accuracy)
```
